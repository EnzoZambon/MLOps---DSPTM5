```python
######## TRATAMIENTO DE NULOS ########

#1. saldo_mora: 0 porque es probable que no tenga mora si no hay dato
df["saldo_mora"] = df["saldo_mora"].fillna(0)

#2. saldo_mora_codeudor: 0 porque es probable que no tenga mora si no hay dato
df["saldo_mora_codeudor"] = df["saldo_mora_codeudor"].fillna(0)

#3. saldo_principal: mediana porque es menos sensible a outliers
df["saldo_principal"] = df["saldo_principal"].fillna(df["saldo_principal"].median())


#4. saldo_total: saldo_principal + saldo_mora porque es la suma de ambos
df["saldo_total"] = df["saldo_total"].fillna(
    df["saldo_principal"] + df["saldo_mora"])


#5. promedio_ingresos_datacredito: mediana porque es menos sensible a outliers
df["promedio_ingresos_datacredito"] = df["promedio_ingresos_datacredito"].fillna(
    df["promedio_ingresos_datacredito"].median())

#6. tendencia_ingresos: moda porque es la categoría más común
df["tendencia_ingresos"] = df["tendencia_ingresos"].fillna(
    df["tendencia_ingresos"].mode()[0])
```

```python
cols = [
    "tipo_credito",
    "plazo_meses",
    "edad_cliente",
    "tipo_laboral",
    "cant_creditosvigentes",
    "huella_consulta",
    "creditos_sectorFinanciero",
    "creditos_sectorCooperativo",
    "creditos_sectorReal",
]

for col in cols:
    plt.figure(figsize=(8,4))

    # Conteo de frecuencias (sin fijar rango)
    freq = df[col].dropna().value_counts().sort_index()

    # Si la variable es categórica (texto), reindexamos para que no queden etiquetas amontonadas
    if df[col].dtype == "object":
        plt.bar(range(len(freq)), freq.values)
        plt.xticks(range(len(freq)), freq.index, rotation=45, ha="right")
        plt.xlabel(col)
    else:
        # Si es numérica, usamos directamente los valores en el eje X
        plt.bar(freq.index, freq.values)
        plt.xlabel(col)

    plt.title(f"Distribución de {col}")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()
```

ANALISIS MULTIVARIABLE

```python
df_mv = df[[
"saldo_total",
"saldo_principal",
"saldo_mora_codeudor",
"creditos_sectorFinanciero",
"creditos_sectorCooperativo",
"creditos_sectorReal",
"promedio_ingresos_datacredito",
"Pago_atiempo"
]].copy()

```

```python
corr_matrix = df_mv.corr().round(2)
corr_matrix
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix, 
    annot=True,        
    fmt=".2f",          
    cmap='coolwarm',     
    center=0,    
    linewidths=0.5,      
    cbar_kws={"shrink": .8} 
)

plt.title("Matriz de Correlación - Análisis Multivariable", fontsize=15)
plt.show()
```
